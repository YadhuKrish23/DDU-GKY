### Percentage of null values in Province/State column id >70 so dropping that column

```python
data=data.drop(columns=['Province/State'])
data
```

```python
data.Date.unique
```

```python
data.Date.nunique
```

```python
data['Date']=pd.to_datetime(data.Date)
data.head()
```

```python
data.info()
```

### Extracting year monyh day from the date column

```python
data['Year']=pd.DatetimeIndex(data.Date).year
data['Month']=pd.DatetimeIndex(data.Date).month
data['Day']=pd.DatetimeIndex(data.Date).day
data.head()
```

### Droping lat,long,date column

```python
data.drop(columns=['Lat','Long','Date'],inplace=True)
```

### Lets convert month in numbers to their corresponding names

```python
data.Month.replace({
     1:'JANUARY',
     2:'FEBRUARY',
     3:'MARCH',
     4:'APRIL',
     5:'MAY',
     6:'JUNE',
     7:'JULY',
     8:'AUGUST',
     9:'SEPTEMBER',
     10:'OCTOBER',
     11:'NOVEMBER',
     12:'DECEMBER'
},inplace=True)
```

```python
data.sample(5)
```

### Rename columns Country/Region to Region

```python
data.rename(columns={'Country/Region':'Region'},inplace=True)
data.head(10)
```

```python
data.rename(columns={'WHO Region':'WHO_Region'},inplace=True)
data.head(1000)
```

## INSIGHTS

### 1. Total deaths from each region

```python
region_death=data.groupby(['Region'])['Deaths'].sum()
region_death
```

### 2. Total active cases from each region

```python
active_cases=data.groupby(['Region'])['Active'].sum()
active_cases.sort_values(ascending=False,inplace=True)
active_cases
```

### 3. No of people who recovered from Canada

```python
data.query('Region=="Central African Republic"')['Recovered'].sum()
```

### 4.Count of people who confitmed positive in january 22

```python
data.query('Month=="January" & Day==22')['Confirmed'].sum()
```

### 5.Mean and sum of persons who confirmed from each region

```python
region_confirmed_mean_sum=data.groupby(['Region'])['Confirmed'].agg(['mean','sum'])
region_confirmed_mean_sum
```

### 6.Average of confirmed persons in each WHO region 

```python
avg_confirmed_WHOregion=data.groupby(['WHO_Region'])['Deaths']
avg_confirmed_WHOregion.mean()
```
